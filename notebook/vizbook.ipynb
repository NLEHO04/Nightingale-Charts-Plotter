```python
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
from PIL import Image

from mplsoccer import PyPizza, add_image, FontManager
```

```python
font_bold = fm.FontProperties(fname="../fonts/BasierCircle-Bold.ttf")
font_medium = fm.FontProperties(fname="../fonts/BasierCircle-Medium.ttf")
font_normal = fm.FontProperties(fname="../fonts/BasierCircle-Regular.ttf")
```

```python
# parameter list
params = [
    "Non-Penalty Goals", "npxG", "npxG per Shot", "\nxA", "Open Play\nSCA",
    "\nPenalty Area\nEntries", "Progressive\nPasses", "Progressive\nCarries",
    "Successful\nDribbles", "\nTouches\nper Turnover", "pAdj\nPress Regains", "Aerials Won"
]

# values for corresponding parameters
values = [99, 99, 87, 51, 62, 58, 45, 40, 27, 74, 77, 73]

# instantiate PyPizza class
baker = PyPizza(
    params=params,                  # list of parameters
    background_color="#EBECE5",
    straight_line_color="#EBECE5",  # color for straight lines,
    straight_line_lw=2,             # linewidth for straight lines
    last_circle_lw=0,               # linewidth of last circle
    other_circle_lw=0,              # linewidth for other circles
)



# plot pizza
fig, ax = baker.make_pizza(
    values,              # list of values
    
    figsize=(12,12),      # adjust figsize according to your need
    
    param_location=107,  # where the parameters will be added
    
    color_blank_space=["#6495ED"]*len(params),
    
    blank_alpha=0.2,
    
    kwargs_slices=dict(
        facecolor="#65adff", edgecolor="#EBECE5",
        zorder=2, linewidth=2
    ),                   # values to be used when plotting slices
    
    kwargs_params=dict(
        color="#222222", fontsize=12,
        fontproperties=font_medium, va="center"
    ),                   # values to be used when adding parameter
    
    kwargs_values=dict(
        color="#101010", fontsize=12,
        fontproperties=font_normal, zorder=3,
        bbox=dict(
            edgecolor="#101010", facecolor="#65adff",
            boxstyle="circle,pad=0.3", lw=1
        )
    )                    # values to be used when adding parameter-values
)

# add title
fig.text(
    0.515, 0.99, "Robert Lewandowski - FC Bayern Munich", size=20,
    ha="center", fontproperties=font_bold, color="#222222"
)

# add subtitle
fig.text(
    0.515, 0.947,
    "Percentile Rank vs Top-Five League Forwards | Season 2020-21\nMinutes: 3476 | Age: 32 years",
    size=15,
    ha="center", fontproperties=font_normal, color="#222222"
)

# # add credits
CREDIT_1 = "Source: Statsbomb via FBRef.com"
CREDIT_2 = "Made Using: mplsoccer\nInspired By: @Worville, @FootballSlices, @somazerofc & @Soumyaj15209314"

fig.text(
    0.925, 0.05, f"{CREDIT_1}", size=9,
    fontproperties=font_medium, color="#222222",
    ha="right"
)

fig.text(
    0.0925, 0.05, f"{CREDIT_2}", size=9,
    fontproperties=font_medium, color="#222222",
    ha="left"
)

fig.savefig("simple.png", dpi=500, bbox_inches="tight", pad_incbhes=0.04)

plt.show()
```
